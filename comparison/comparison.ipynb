```python
import pandas as pd
```

```python
orig = pd.read_csv('original_data_w_metrics.tsv', sep='\t')
synth = pd.read_csv('df_with_3_metrics.tsv', sep='\t')
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

def analysis(orig_df, synthetic_df):
  sns.set_style("whitegrid")
  fig, axs = plt.subplots(1,3, figsize=(12, 5))
  for i,col in enumerate(orig_df.columns):
    sns.kdeplot(data=[orig_df[col].values,synthetic_df[col].values],legend=False,ax=axs[i],fill=True)
    axs[i].set_title(f'{col} scores distribution')
    axs[i].legend(labels=["synthetic", "original"], loc='center left') #  . 
    #  original         synthetic, ,   . 
    #   .
  plt.tight_layout()
  plt.show()

analysis(
    orig[['faithfulness', 'answer_relevancy', 'context_precision']],
    synth[['faithfulness', 'answer_relevancy', 'context_precision']]
) 
```

```python
def hist_comp(orig_df, synthetic_df):
  sns.set_style("whitegrid")
  fig, axs = plt.subplots(1,3, figsize=(12, 5))
  for i,col in enumerate(orig_df.columns):
    sns.histplot(data=[orig_df[col].values,synthetic_df[col].values],legend=False,ax=axs[i],fill=True)
    axs[i].set_title(f'{col} scores distribution')
    axs[i].legend(labels=["synthetic", "original"], loc='best') #   .
  plt.tight_layout()
  plt.show()

hist_comp(
    orig[['faithfulness', 'answer_relevancy', 'context_precision']],
    synth[['faithfulness', 'answer_relevancy', 'context_precision']]
) 
```
